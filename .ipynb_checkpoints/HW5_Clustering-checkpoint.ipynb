````python
import warnings
from scipy.spatial.distance import cdist


class KMeans():
    def __init__(self, n_clusters, max_iter=100):
        self.k = n_clusters
        self.max_iter = max_iter
    
    def cluster(self, X):
        """
        Do K-Means clustering
        
        Parameters
        ----------
        X: np.ndarray
            Normalized input features, shape (n_samples, n_features)
        """
        centroids = np.random.random((self.k, X.shape[1]))
        assignments = np.zeros(X.shape[0])
        new_assignments = self.create_new_assignments(centroids, X)
        
        # restart if run into bad initialization
        # Comment out this part for Q1.(d)
        if len(np.unique(new_assignments)) < self.k:
            return self.cluster(X)
        
        n_iters = 1
        while (new_assignments != assignments).any() and n_iters < self.max_iter:
            # compute centroids
            centroids = np.array([X[new_assignments == k].mean(axis=0) for k in range(self.k)])
            assignments = new_assignments
            # call the self.create_new_assignments
            new_assignments = self.create_new_assignments(centroids, X)
            if len(np.unique(new_assignments)) < self.k:
                warnings.warn('Warning: At least one centroid vanishes')
            n_iters += 1
            if n_iters == self.max_iter:
                warnings.warn("Warning: Maximum number of iterations reached!")
        return new_assignments, centroids
            
        
    def create_new_assignments(self, centroids, X):
        """
        Assign each datapoint to its nearest centroid.
        
        Parameters
        ----------
        centroids: np.ndarray
            Centroid of each cluster, shape (n_clusters, n_features)
        X: np.ndarray
            Input features, shape (n_samples, n_features)
        
        Returns
        ------
        new_assignments: np.ndarray
            Index of cluster which each data point belongs to, shape (n_samples,)
        """
        # Compute the distances that stores the Eucledian distances between each datapoints and the centroid
        # shape (n_samples, n_clusters)
        # Hint: you can use cdist in scipy.spatial.distance. 
        # cdist doc: https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.cdist.html
        distances = cdist(X, centroids, 'euclidean')
        new_assignments = np.argmin(distances, axis=-1)
        return new_assignments
````

````python
def validate(y_pred, y_true):
    """
    Validate the results of clustering
    
    Parameters
    ---------
    y_pred: np.ndarray
        Predicted clusters
    y_true: list or np.ndarray
        Ground truth clusters
    
    Example
    ------
    >>> validate(model.cluster(features), df['type'].values)
    """
    # correct classification
    compounds = np.unique(y_true) # should be ['amide','phenol','ether'] for compounds dataset
    clusters = [np.where(y_true == c)[0] for c in compounds]
    pred_class = np.unique(y_pred)
    
    #remove -1 for noise point in DBSCAN
    pred_class = np.delete(pred_class, np.where(pred_class == -1))
    assert len(pred_class) == len(compounds), \
        f'Predicted clusters has less or more than {len(compounds)} classes: {pred_class}'
    for i in range(3):
        #loop over solutions
        counts = []
        scores = []
        for j in range(3):
            # loop over clusters of true assignments
            sol_i = np.where(y_pred == pred_class[i])[0]
            counts.append(len(np.intersect1d(sol_i, clusters[j])))
            scores.append(counts[-1] / len(clusters[j]))
        idx = np.argmax(scores)
        print(
            f'Class {pred_class[i]} - {compounds[idx]}: {counts[idx]} out of {len(clusters[j])}'
            ' are classified correctly'
        ) 
````

````python
class KMeansNoRestart(KMeans):
    def cluster(self, X):
        """
        Do K-Means clustering without restarting on bad initialization
        
        Parameters
        ----------
        X: np.ndarray
            Normalized input features, shape (n_samples, n_features)
        """
        centroids = np.random.random((self.k, X.shape[1]))
        assignments = np.zeros(X.shape[0])
        new_assignments = self.create_new_assignments(centroids, X)
        
        n_iters = 1
        while (new_assignments != assignments).any() and n_iters < self.max_iter:
            centroids = np.array([X[new_assignments == k].mean(axis=0) for k in range(self.k)])
            assignments = new_assignments
            new_assignments = self.create_new_assignments(centroids, X)
            if len(np.unique(new_assignments)) < self.k:
                warnings.warn('Warning: At least one centroid vanishes')
            n_iters += 1
            if n_iters == self.max_iter:
                warnings.warn("Warning: Maximum number of iterations reached!")
        return new_assignments, centroids
````

````python
ntrials = 50
count = 0
k = 4
for i in range(ntrials):
    kmeans = KMeansNoRestart(n_clusters=k) # Comment out the re-init code in KMeans
    results, centroids = kmeans.cluster(com_df_scaled[features].values)
    if len(np.unique(results)) == k:
        count += 1
print(f'Out of {ntrials} number of KMeans trials, only {count} gives {k} clusters')
````

### Description: 
<br> Based on the results when running KMeans with K=4 multiple times without allowing the algorithm to restart on bad initializations, we may find that not all trials result in 4 distinct clusters. **This could be due to the random initialization leading to poor starting conditions, such as two centroids being initialized too close to each other and capturing the same cluster, or a centroid being placed too far from any data points and ending up with no points assigned to it.**
#### Problems:
* **Random Initialization Problem:** KMeans is sensitive to the initial placement of centroids. Random initialization can sometimes place centroids in suboptimal locations, which leads to poor clustering results. In some cases, one or more clusters may have very few or no points assigned to them, especially if the random initialization places a centroid far from the data points.
* **Local Minima:** KMeans aims to minimize the within-cluster variance, but it can get stuck in a local minimum rather than the global minimum. This means that the algorithm may converge to different solutions in different runs, some of which may be suboptimal.
#### Solutions: 
* Improved Initialization: Using methods like k-means++ for smarter centroid initialization, which tends to place centroids far from each other and reduces the chances of empty clusters.
* Multiple Runs with Selection Criteria: Running KMeans multiple times and selecting the best run based on a certain criterion (e.g., the lowest within-cluster sum of squares), known as the ensemble method.

### Optimization 1- using k-means++

````python
def kmeans_plus_plus(X, n_clusters):
    """
    Initialize the centroids according to the k-means++ method for KMeans clustering.

    Parameters
    ----------
    X : np.ndarray
        The input samples, shape (n_samples, n_features).
    n_clusters : int
        The number of centroids to generate.

    Returns
    -------
    centroids : np.ndarray
        The initialized centroids, shape (n_clusters, n_features).
    """
    n_samples, n_features = X.shape

    # Choose the first centroid randomly from the data points
    centroids = np.zeros((n_clusters, n_features))
    initial_idx = np.random.choice(n_samples)
    centroids[0] = X[initial_idx]

    # Compute distances from the first centroid chosen to all the other data points
    distances = np.linalg.norm(X - centroids[0], axis=1)

    for i in range(1, n_clusters):
        # Select the next centroid randomly weighted by the square of distances
        probabilities = distances ** 2
        probabilities /= probabilities.sum()
        centroid_idx = np.random.choice(n_samples, p=probabilities)
        centroids[i] = X[centroid_idx]

        # Update the distances based on the new centroid, if they are smaller
        new_distances = np.linalg.norm(X - centroids[i], axis=1)
        distances = np.minimum(distances, new_distances)

    return centroids

# Now let's modify the KMeans class to use k-means++ initialization
class KMeans_Opt1:
    def __init__(self, n_clusters, max_iter=100, init='kmeans++'):
        self.k = n_clusters
        self.max_iter = max_iter
        self.init = init
    
    def cluster(self, X):
        if self.init == 'kmeans++':
            centroids = kmeans_plus_plus(X, self.k)
        else:
            centroids = np.random.random((self.k, X.shape[1]))
        assignments = np.zeros(X.shape[0])
        new_assignments = self.create_new_assignments(centroids, X)
        
        n_iters = 1
        while (new_assignments != assignments).any() and n_iters < self.max_iter:
            centroids = np.array([X[new_assignments == k].mean(axis=0) for k in range(self.k)])
            assignments = new_assignments
            new_assignments = self.create_new_assignments(centroids, X)
            if len(np.unique(new_assignments)) < self.k:
                warnings.warn('Warning: At least one centroid vanishes')
            n_iters += 1
            if n_iters == self.max_iter:
                warnings.warn("Warning: Maximum number of iterations reached!")
        return new_assignments, centroids
            
    def create_new_assignments(self, centroids, X):
        distances = cdist(X, centroids, 'euclidean')
        new_assignments = np.argmin(distances, axis=1)
        return new_assignments
````

````python
ntrials = 50
count = 0
k = 4
for i in range(ntrials):
    kmeans = KMeans_Opt1(n_clusters=k) # Comment out the re-init code in KMeans
    results, centroids = kmeans.cluster(com_df_scaled[features].values)
    if len(np.unique(results)) == k:
        count += 1
print(f'Out of {ntrials} number of KMeans trials, only {count} gives {k} clusters')
````

````python
# Number of clusters in labels, ignoring noise if present
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

# Calculate the number of core points
n_core_points = len(cores)

# Calculate the number of border points, the number of points that is not core points and noise points
# Firstly, ~core_samples_mask turns the points that originally are False to True, and then eliminate the labels is not -1
n_border_points = sum(~core_samples_mask & (labels != -1)) 

# Print out the results
print(f"Estimated number of clusters: {n_clusters}")
print(f"Number of core points: {n_core_points}")
print(f"Number of border points: {n_border_points}")
print(f"Number of noise points: {n_noise}")
````

````python
from sklearn import datasets

def moon_dataset(n_samples=1500):
    """
    Generate datasets. We choose the size big enough to see the scalability
    of the algorithms, but not too big to avoid too long running times
    
    Parameters
    ---------
    n_samples: int
        Number of data points
    
    Returns
    ------
    X: np.ndarray
        Data points, shape (n_samples, 2)
    y: np.ndarray
        Clusters that each point belongs to, shape (n_samples)
    """
    np.random.seed(0)
    n_samples = 1500
    noisy_circles = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05)
    noisy_moons = datasets.make_moons(n_samples=n_samples, noise=0.05)
    X, y = noisy_moons
    return X, y

X, y = moon_dataset()

# Standardize the data
scaler = StandardScaler()
X_scaled_M = scaler.fit_transform(X)

plt.scatter(X[:, 0], X[:, 1], s=10, c=y, cmap='cool')
````

````python
# Define 'eps' and 'min_smaples' for each runs
dbscan_runs = [
    {'eps': 0.1, 'min_samples': 5},
    {'eps': 0.1, 'min_samples': 10},
    {'eps': 0.25, 'min_samples': 5}
]
# Plotting the results
fig, axes = plt.subplots(1, 3, figsize=(18, 4))
for i, params in enumerate(dbscan_runs):
    db = DBSCAN(eps=params['eps'], min_samples=params['min_samples'])
    db.fit(X_scaled_M)

    axes[i].scatter(X_scaled_M[:, 0], X_scaled_M[:, 1], c=db.labels_, s=10, cmap='cool')
    axes[i].set_title(f'DBSCAN Run {i+1}')

    cores = db.core_sample_indices_
    labels = db.labels_
    core_samples_mask = np.zeros_like(labels, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True 
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)
    n_core_points = len(cores)
    n_border_points = sum(~core_samples_mask & (labels != -1)) 

    print(f"DBSCAN Run {i+1} (eps: {dbscan_runs[i]['eps']}, min_samples: {dbscan_runs[i]['min_samples']})")
    print(f"Estimated number of clusters: {n_clusters}")
    print(f"Number of core points: {n_core_points}")
    print(f"Number of border points: {n_border_points}")
    print(f"Number of noise points: {n_noise}\n")

plt.show()
````

#### The effectiveness of DBSCAN in identifying the moon shapes depends significantly on the choice of its two main parameters:

* eps (epsilon): The maximum distance between two samples for one to be considered as in the neighborhood of the other. If eps is chosen too small, the crescent shapes won't be recognized as single clusters. If it's too large, the two crescents might be merged into one cluster.
* min_samples: The number of samples in a neighborhood for a point to be considered as a core point. This includes the point itself.

* No Need for a Priori Cluster Number: KMeans requires the number of clusters to be specified in advance. However, DBSCAN does not require this and can determine the number of clusters based on the data provided, which can be an advantage when the number of clusters is not known beforehand.

````python
# If you feel the code is slow in Q3(d), use numba to accelerate it
import numba

@numba.njit()
def cost_func(features, centroids, clusters):
    """
    Parameters
    ----------
    features: np.ndarray
        Feature matrix, shape (n_samples, n_features). In the compound dataset, n_features = 4.
    centroids: np.ndarray
        Centroids of each cluster, shape (n_cluster, n_features). In this question, n_cluster = 3.
    clusters: np.ndarray
        Cluster which each datapoint belongs to. shape (n_samples,)
    """
    cost = 0
    for i in range(len(features)):
        # Calculate Euclidean distance from the point to its assigned centroid
        distance = np.linalg.norm(features[i] - centroids[clusters[i]])
        # Accumulate the squared distance to the cost
        cost += distance ** 2
    return cost
````

````python
# Extract the features and convert to numpy array for the cost function calculation
features = com_df_norm.iloc[:, :4].values

# Convert the centroids to numpy array
centroids = centroids.values

# Map 'Start assignment' to its corresponding index (0, 1, 2)
clusters = com_df_norm['Start assignment'].values

# Calculate the cost for the initial categorization
initial_cost = cost_func(features, centroids, clusters)
initial_cost
````

## (d)

*For debugging*: 
+ Cooling schedule should be like: `array([500. , 499.5, 499.0005, ..., 3.37065784, 3.36728719, 3.3639199])`
+ Cost function can drop to around 540 (or even smaller).

````python
def SA(features, cluster, schedule, report_interval=100):
    """
    Simulated Annealing for clustering
    
    Parameters
    ----------
    features: numpy.ndarray
        Feature matrix. shape (n_samples, n_features)
    cluster: numpy.ndarray
        Initial guess of clusters. shape (n_samples,)
    schedule: numpy.ndarray
        Cooling schedule.
    report_interval: int
        Interval of steps to report result
    
    Returns
    ------
    res: dict
        Result containing the best clustering with the cost function value
    """
    # calculate the centroids
    centroids = np.array([features[cluster == i].mean(axis=0) for i in np.unique(cluster)])
    # calculate the cost function
    cluster_eval = cost_func(features, centroids, cluster)
    
    best_cluster = cluster.copy()
    lowest_eval = cluster_eval

    for step, temp in enumerate(schedule):
        if step % report_interval == 0:
            print(step, temp, lowest_eval)
        for n in range(features.shape[0]):
            trial = cluster.copy()
            # Do a Metroplis step by swtiching a sample to differenct cluster
            choices = [0, 1, 2]
            choices.remove(cluster[n]) # remove the current cluster
            rand_choice = np.random.choice(choices) # choose a different cluster
            trial[n] = rand_choice # switch the cluster
            
            # calculate the centroids & cost function
            trial_centroids = np.array([features[trial == i].mean(axis=0) for i in np.unique(trial)])
            trial_eval = cost_func(features, trial_centroids, trial)
            # Metroplis rule
            if trial_eval < cluster_eval or np.random.rand() < np.exp((cluster_eval - trial_eval) / temp):
                cluster = trial
                cluster_eval = trial_eval
                if trial_eval < lowest_eval:
                    best_cluster = cluster.copy()
                    lowest_eval = cluster_eval
                    
    return {"solution": best_cluster, "evaluation": lowest_eval}
````

````python
def geometric_cooling(init_temp, num_steps, alpha):
    return np.array([init_temp * (alpha ** i) for i in range(num_steps)])
````

````python
# Initialize variables
init_temp = 500
num_steps = 5000
alpha = 0.999
schedule = geometric_cooling(init_temp, num_steps, alpha)
pd.DataFrame(schedule)
````

````python
# Plotting geometric cooling schedule
plt.plot(schedule, color='violet')
plt.title('Geometric Cooling')
plt.xlabel('Iteration')
plt.ylabel('Temperature (K)')
````

````python
# Perform the simulated annealing
for i in range(3):
    print(f"## Trial {i} ##")
    res = SA(features, clusters, schedule)
    
    # print the members of the phenol, ether, and amide as part of each cluster
    phenol_count = np.sum(res['solution'] == 0)
    ether_count = np.sum(res['solution'] == 1)
    amide_count = np.sum(res['solution'] == 2)
    print(f"Phenol members: {phenol_count}")
    print(f"Ether members: {ether_count}")
    print(f"Amide members: {amide_count}")
    
    validate(res['solution'], com_df['type'].values)
````

### Description 
#### KMeans has provided a perfect classification for phenols and substantially improved classification for ethers and amides compared to SA. The reasons KMeans may perform better than SA for this dataset include:
* Data Distribution: KMeans works well when the clusters are spherical and well separated. If the compounds dataset naturally forms such clusters, KMeans would naturally perform well.
* Algorithm Suitability: KMeans is a centroid-based algorithm, which is generally faster and more efficient for large datasets with clear cluster separation. If the compounds form distinct groups based on their features, KMeans can capitalize on that structure.
